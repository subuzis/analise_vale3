```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")  ###estilo dos graficos
```

```python
#criando o Data Frame
df = pd.read_csv("all_ibovespa.csv", delimiter=',')
df
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
###Criando Data Frame Vale

df_vale = df[df['sigla_acao'] == 'VALE3']

#verificando a data inicial das consultas
df_vale.head(10)
```

```python
##verificar último dia do pregão consultado
df_vale.tail()
```

```python
#Consultando linhas sem valores
df_vale.isnull().sum()
```

```python
#verificar o tipo do arquivo
df_vale.dtypes

```

```python
#Mudar o tipo data
df_vale['data_pregao'] = pd.to_datetime(df_vale['data_pregao'], format='%Y-%m-%d')
```

```python
#verificar o tipo do arquivo
df_vale.dtypes
```

```python
#criando novos campos de médias móveis

df_vale['MM5D'] = df_vale['preco_fechamento'].rolling(5).mean()
df_vale['MM21D'] = df_vale['preco_fechamento'].rolling(21).mean()
```

```python
df_vale
```

```python
#verificar o tipo do arquivo
df_vale.dtypes
```

```python
##Alterar os valores NaN por 0
df_vale["MM5D"].fillna(0, inplace=True)
df_vale["MM21D"].fillna(0, inplace=True)
```

```python
##Validando alteração anterior
df_vale
```

```python
##Informações gerais sobre o preço de fechamento
df_vale["preco_fechamento"].describe()
```

```python
#Gráfico de Boxplot preço de fechamento
plt.boxplot(df_vale["preco_fechamento"]);
plt.savefig("BoxPlot.png")
```

```python
#Maior preço de fechamento por ano
df_vale.groupby(df_vale["data_pregao"].dt.year)["preco_fechamento"].max()
```

```python
##Gráfico de Maior preço de fechamento por ano
df_vale.groupby(df_vale["data_pregao"].dt.year)["preco_fechamento"].max().plot(title="Preço x Ano", marker="*")
plt.xlabel("Ano")
plt.ylabel("Preço");
plt.savefig("Maio preço de fechamento por ano.png")
```

```python
#Maior preço de fechamento por mes
df_vale.groupby(df_vale["data_pregao"].dt.month)["preco_fechamento"].max()
```

```python
#Verificando exatamente em qual dia ocorreu o maior preço de fechamento do Ativo
df_vale[df_vale["preco_fechamento"] == 118.72]
```

```python
#Menor preço de fechamento por ano
df_vale.groupby(df_vale["data_pregao"].dt.year)["preco_fechamento"].min()
```

```python
##Gráfico de Menor preço de fechamento por ano
df_vale.groupby(df_vale["data_pregao"].dt.year)["preco_fechamento"].min().plot(title="Preço x Ano", marker="*")
plt.xlabel("Ano")
plt.ylabel("Preço");
plt.savefig("Menor preço de fechamento por ano.png")
```

```python
#Menor preço de fechamento por mes
df_vale.groupby(df_vale["data_pregao"].dt.month)["preco_fechamento"].min()
```

```python
#Verificando exatamente em qual dia ocorreu o menor preço de fechamento do Ativo
df_vale[df_vale["preco_fechamento"] == 8.60]
```

```python
##salvar sem o index do Python
df_vale.to_csv("vale3_hist.csv", index=False)
```
